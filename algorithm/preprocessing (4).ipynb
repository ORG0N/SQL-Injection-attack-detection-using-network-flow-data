```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
```

```python

def new_dataset(dataframe): # function to remove "Label" column
    print("Removing Label Column....\n")
    lc=dataframe["Label"]                        #Target column print
    new_dataframe=dataframe.drop("Label",axis=1) # other than Label(Target)
    new_columns=dataframe.columns.drop("Label") # Head return
    return new_columns,lc,new_dataframe
    
def MinScale_Train(dataframe):
    print("MinScale Normalization for Train....\n")
    heading=dataframe.columns
    class_call=MinMaxScaler()
    scaled=class_call.fit_transform(dataframe)
    scaled=pd.DataFrame(scaled,columns=heading)
    return scaled
    
def MinScale_Test(dataframe,test_df):
    print("MinScale Normalization for Test.....\n")
    heading=dataframe.columns
    class_call=MinMaxScaler()
    scaled_test=class_call.fit_transform(test_df)
    scaled=pd.DataFrame(scaled_test,columns=heading)
    return scaled
    
def MaxScale_Train(dataframe):
    print("MaxAbs Normalization Train:")
    heading=dataframe.columns
    class_call=MaxAbsScaler()
    trained=class_call.fit_transform(dataframe)
    trained=pd.DataFrame(trained,columns=heading)
    return trained
    
def MaxScale_Test(dataframe,test_df):
    print("MaxAbs Normalization Test....\n")
    heading=dataframe.columns
    scale=MaxAbsScaler()
        #scale_trained=self.maxabsNormal_train(dataframe)
    scale_test=scale.fit_transform(test_df)
    scaled_test=pd.DataFrame(scale_test,columns=heading)
    print(scale_test)
    
def Robust_Train(dataframe):
    print("Robust Normalization Train...\n")
    heading=dataframe.columns
    class_call=RobustScaler()
    trained=class_call.fit_transform(dataframe)
    trained=pd.DataFrame(trained,columns=heading)
    return trained
    
def Robust_Test(dataframe,test_df):
    print("Robust Normalization Test...\n")
    heading=dataframe.columns
    scale=RobustScaler()
        #scale_trained=self.robust_train(dataframe)
    scale_test=scale.fit_transform(test_df)
    scaled_test=pd.DataFrame(scale_test,columns=heading)
    print(scale_test)
    
def Variance_cal(dataframe,limit=0,list=["Label"]):
    print("Calculating Variance....\n")
    for column in dataframe:
        f=False
        for itr in list:
            if itr==column:
                f=True
        if f==False:
            variance=dataframe[column].var()
            print("Column:",column,"\nVariance:",variance)  # Reference
            if variance==limit:
                dataframe=dataframe.drop(column,axis=1)
        
data=pd.read_csv("E:\\jupyter\\dataset\\GivenDataset.csv")
head,target,new_dataframe=new_dataset(data)
print(new_dataframe,end="\n")
print(target,end="\n")
print(head,end="\n")

print(end="\n")
scaled_value=MinScale_Train(new_dataframe)
scaled_value["Label"]=target
shuffled_scaled_value=scaled_value.sample(frac=1)
shuffled_scaled_value.to_csv("E:\\jupyter\\dataset\\MinScale\\minscaleTrain.csv")
print(shuffled_scaled_value)


print(end="\n")
scaled_value=MaxScale_Train(new_dataframe)
scaled_value["Label"]=target
shuffled_scaled_value=scaled_value.sample(frac=1)
shuffled_scaled_value.to_csv("E:\\jupyter\\dataset\\MinScale\\maxscaleTrain.csv")
print(shuffled_scaled_value)

print(end="\n")
scaled_value=Robust_Train(new_dataframe)
scaled_value["Label"]=target
shuffled_scaled_value=scaled_value.sample(frac=1)
shuffled_scaled_value.to_csv("E:\\jupyter\\dataset\\MinScale\\robustscaleTrain.csv")
print(shuffled_scaled_value)

print(end="\n")
Variance_cal(data)
```
